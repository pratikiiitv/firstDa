```r
x<-read.table('cotton_strength.txt')
```

```r
names(x)
```

```r
plot(x$V1,x$V2)
```

```r
summary(x)
```

```r
hist(x$V2)
```

```r
v<-x$V1; v2<-v*v;
```

```r
u<-x$V2
```

```r
qm<-lm(u~v+v2)
```

```r
summary(qm)
```

```r
vs<-seq(5,40,1);
```

```r
us<-predict(qm,list(v=vs, v2=vs*vs))
```

```r
plot(v,u); lines(vs,us)
```

```r
x<-read.csv('internet_traffic.csv')
```

```r
library(tidyr)
```

```r
library(ggplot2)
```

```r
xt<-gather(x,key=weekday,value=inTr,-X)
```

```r
ggplot(xt,mapping=aes(x=weekday,y=inTr)) + geom_boxplot() + geom_jitter(alpha=.3)
```

```r
library(dplyr)
```

```r
summarize(group_by(xt,weekday),mean(inTr))
```

```r
n<-500; t<-seq(1,n,1); o1<-rnorm(n); o2<-2+rnorm(n);
```

```r
tsd<-data.frame(id=c(rep(1,n),rep(2,n)),day=c(t,t),ts=c(o1,o2))
```

```r
head(tsd)
```

```r
ggplot(tsd,aes(x=day,y=ts,group=id,color=id))+geom_line()
```

```r
ggplot(tsd,aes(day,ts,color=id))+geom_line()+facet_wrap(vars(id))
```

```r
ggplot(tsd,aes(id,ts,group=id))+geom_boxplot()
```

```r
ggplot(tsd,aes(id,ts,color=id,group=id))+geom_boxplot()+geom_jitter(alpha=0.3)
```
