Problem : 1
---------------
Suppose that an engineer encounters data from a manufacturing
process in which 100 items are observed and 10 are found to be
defective. It is expected and anticipated that occasionally there
will be defective items. However, it has been determined that in
the long run, the company can only tolerate 5% defects. What
would be the engineer’s reaction?

Let us recall Binomial probability mass function:
P(x, N, pe) = x_C_N pe^x (1-pe)^(N-x)

Given that probability of faulty event is pe and events are all independent to each other, the probability there are x faulty events in a sequence of N event

p, q, d, r (distribution)
p - probability density function
d - cumulative distribution function
r - random variate/ samples
q - quantile

```r
n<-10; N<-100; pe<-0.05;
k<-seq(0,N,1); Pk<-dbinom(k,N,pe); 
```

```r
plot(k,Pk); lines(k,Pk);
```

Probability that the number of defective items produced are 10 or more.
P(k >= 10) = 1 - P(k<10) = 1 - P(k<=9)
